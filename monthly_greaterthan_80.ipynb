```python
import pandas as pd
data = pd.read_csv("fire_archive_M-C61_2014 - Copy.csv")

# 2. Convert date
data["acq_date"] = pd.to_datetime(data["acq_date"])

# 3. Filter confidence >= 80
conf_lvl = data[data["confidence"] >= 80].copy()

# 4. Extract month
conf_lvl["month"] = conf_lvl["acq_date"].dt.month

# 5. Count per month > 80
monthly_count = conf_lvl.groupby("month").size().reset_index(name="fire_count")
print(monthly_count)
```
